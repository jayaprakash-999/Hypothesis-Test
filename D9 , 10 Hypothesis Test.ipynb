# One tail test

One-sample and one tail t-tests

Ex. An outbreak of Salmonella-related illness was attributed to ice cream produced
at a certain factory. Scientists measured the level of Salmonella in 9 randomly
sampled batches of ice cream. The levels (in MPN/g) were

0.593, 0.142, 0.329, 0.691, 0.231, 0.793, 0.519, 0.392, 0.418

Is there evidence that the mean level of Salmonella in the ice cream is greater than 0.3 MPN/g?

ice cream is greater than 0.3 MPN/g

Ice cream H0 : Mean <0.3   ;
          H1 : mean >0.3

````python
import pandas as pd
data=pd.Series([0.593, 0.142, 0.329, 0.691, 0.231, 0.793, 0.519, 0.392, 0.418])
````

````python
stats.ttest_1samp(data,0.3)
````

````python
#t cal =2.2050588385131595,
# pvalue=0.05853032968489765,
# df=8    (n-1)
````

````python
# This gives two tail test p value
# so divide by 2 to get final p-value
# P value at one tail test divide by 2
# p = [1] = 0.05853032968489765

from scipy import stats
p=stats.ttest_1samp(data,0.3)[1]
p_value= p/2
p_value
````

# Two Tail test

Two-sample t-tests
Ex. 6 subjects were given a drug (tfeatment group) and an additional 6 subjects a placebo (control group). Their reaction time to a stimulus was measured (in ws). We want to perform a two-sample t-test for comparing the means of the treatment and control groups.


Coptrol : 91, 87, 99, 77,88, 91
Treat :101, 110, 103, 93, 99, 104   (assume)

````python
Control=pd.Series([91, 87, 99, 77, 88, 91])
Treat =pd.Series([101, 110, 103, 93, 99, 104])
````

````python
stats.ttest_ind(Control,Treat)
````

````python
#    H0   μ1 = μ 2
#    H1   μ1 != μ2

"""
p = 0.006 < 0.05
reject H0
Accept H1
μ1 != μ2
there is a signficant differnece betewween the two groups
control and treat group are statically significantly differce """
````

# 2 Proportion Test

The difference in population proportion needs t-test. Also, the population follows a binomial distribution here. We can just pass on the two population quantities with the appropriate binomial distribution parameters to the t-test function.

We can use the ttest_ind() function from Statsmodels. The function returns three values: (a) test statistic, (b) p-value of the t-test, and (c) degrees of freedom used in the t-test

Data Given:
n1 = 247
p1 = 0.37

n2=308
p2 = 0.39

92 students = 37% of students got a job out of 247
 students = 39% of students not got a job out of 247

````python
import numpy as np
````

````python
#Data:

n1 = 247
p1 = .37

n2 = 308
p2 = .39
````

````python
population1 = np.random.binomial(1, p1, n1)
population2 = np.random.binomial(1, p2, n2)
````

````python
population1
# 1 Represnt got a job random assumption
````

````python
population2
# 1 Represnt Not got a job random assumption
````

````python
from scipy import stats
stats.ttest_ind(population1, population2)
````
